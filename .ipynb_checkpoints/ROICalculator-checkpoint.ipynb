```python
from IPython.display import clear_output

class ROICalculator():
    
    def __init__(self,est_mortgage = 1000):
        print("DENVER AREA RENTAL PROPERTY ROI CALCULATOR\n=======================================")
        self.est_mortgage = est_mortgage
        self.price = input("Please enter the purchase amount of the property in USDollars: \n$")
        self.units = input("Please enter the number of rental units on the property (1 for single family home): \n")
        try:
            self.down_pymt = input(f"Please enter the amount of down-payment made on the loan for the property.\n (Typically 5%-20% or ${int(self.price) * .05:.0f}-${int(self.price) * .20:.0f} for this property: \n$)")
        except:
            print("Please only enter whole numbers without commas or $ signs")
            not_kept = input(f"Press enter to quit")
        self.loan_term = input("Please enter the number of years for repayment in the loan terms. Usually 15 or 30: \n")
        try:
            if int(self.loan_term) <= 15:
                self.est_mortgage = .006 * (int(self.price) - int(self.down_pymt))
            else:
                self.est_mortgage = .004 * (int(self.price) - int(self.down_pymt))
        except:
            print("Please only enter whole numbers without commas or $ signs")
            not_kept = input(f"Press enter to quit")
    
    def get_income(self):
        print("INCOME\n==================")
        self.rent = input('Enter the total amount of all the rent you collect from the tenants per month: \n$')
        laundry = input('Enter the total amount of laundry fees you collect from the tenants each month: \n$')
        storage = input('Enter the total amount of storage fees you collect from the tenants each month: \n$')
        misc = input('Enter the total amount of miscelaneous fees (such as trash removal service, parking ports, etc) collected from the tenants each month: \n$')
        try:
            self.income = int(self.rent) + int(laundry) + int(storage) + int(misc)
            return self.income, self.rent
        except:
            print("Please only enter whole numbers without commas or $ signs")
            not_kept = input(f"Press enter to quit")
            clear_output()
            self.get_income()
    
    def get_expenses(self,rent):
        self.rent = rent
        print("EXPENSES\n=================")
        tax = input(f"Enter the monthly taxes you will pay for this property (typically around ${(int(self.price) * 0.005)/12:.0f}): \n$")
        insurance = input("Enter the monthly insurance you will pay for this property (typically around $85): \n$")
        utilities = input("Enter the cost for utilities for the property (Enter 0 if paid by the tenant): \n$")
        hoa = input("Enter the cost for monthly HOA fees (enter 0 for none): \n$")
        lawn_snow = input("Enter the monthly expenditure for lawn care and snow removal (enter 0 if this is the tenant's responsibility): \n$")
        vacancy = input(f"It is recommended that you set aside 5% of rent charged to defray the cost of vacancies, ${int(self.rent) * .05:.0f}\nEnter the amount to set aside: \n$")
        repairs = input(f"It is recommended that you set aside $100 per month per unit to cover typical repairs, ${int(self.units) * 100:.0f})\nEnter he amount to set aside: \n$")
        capEx = input(f"It is recommended that you set aside 5% of rent per month for capital expenditures (Large item replacements such as roof, furnace, siding) ${int(self.rent) * .05:.0f}\n Enter he amount to set aside: \n$")
        p_manager = input("Enter the amount you pay your property manager each month ($0 if none): \n$")
        mortgage = input(f"Enter the amount of your monthly mortgage payment (typically ${int(self.est_mortgage):.0f} for this property): \n$")
        try:
            self.expenses = int(tax) + int(insurance) + int(utilities) + int(hoa) + int(lawn_snow) + int(vacancy) + int(repairs) + int(capEx) + int(p_manager) + int(mortgage)
            return self.expenses
        except:
            print("Please only enter whole numbers without commas or $ signs")
            not_kept = input(f"Press enter to quit")
            clear_output()
            self.get_expenses()
    
    def find_cashflow(self,income,expenses):
        self.income = income
        self.expenses = expenses
        self.cashflow = self.income - self.expenses
        print("CASH FLOW\n===================")
        print(f"You're estimated cash flow is ${self.cashflow:.0f} per month")
        not_kept = input(f"\nPress enter to continue")
        return self.cashflow
    
    def find_cash_on_cash(self,cashflow):
        self.cashflow = cashflow
        print("CASH ON CASH RETURN ON INVESTMENT\n=========================================")
        closing = input("Enter the amount of the closing costs for the loan on the property: \n$")
        rehab = input("Enter the amount of you spent or plan to spend to repair and remodel the property: \n$")
        misc_other = input("Enter the USDollar amount of any other miscelaneous expenditures you made or will make on the property: \n$")
        try:
            total_invest = int(self.down_pymt) + int(closing) + int(rehab) + int(misc_other)
            self.cash_on_cash = ((12 * self.cashflow)/total_invest)*100
            print("====================================================")
            print(f"The cash-on-cash return for this investment is {self.cash_on_cash:.2f}%")
            print("====================================================")
        except:
            print("Please only enter whole numbers without commas or $ signs")
            not_kept = input(f"Press enter to quit")
            clear_output()
            self.find_cash_on_cash()
    

test_case = ROICalculator()
clear_output()
income, rent = test_case.get_income()
clear_output()
expenses = test_case.get_expenses(rent)
clear_output()
cashflow = test_case.find_cashflow(income, expenses)
clear_output()
test_case.find_cash_on_cash(cashflow)


```
